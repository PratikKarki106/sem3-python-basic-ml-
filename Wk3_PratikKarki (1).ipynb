# **Exploratory data analysis with group-by method**

**What percent of the total fare revenue came from each passenger class**

Calculate the total fare paid across all passengers in the titanic dataset

```python
total_fare = df['Fare'].sum()
print("The total fare is: ",total_fare)
```

Subset for passengers in first class (Pclass is 1) and calculate their total fare.

```python
first_class_fare = df[df['Pclass'] == 1]['Fare'].sum()
print(f"The total fare in first class is: {first_class_fare}")
```

Do the same for second class (Pclass is 2) and third class (Pclass is 3).

```python
second_class_fare = df[df['Pclass'] == 2]['Fare'].sum()
third_class_fare = df[df['Pclass'] == 3]['Fare'].sum()
print("The total fare in second class is: ",second_class_fare)
print("The total fare in third class is: ",third_class_fare)
```

Combine the fare totals from first, second, and third classes into a list.

```python
first_class_fare = df[df['Pclass'] == 1]['Fare'].sum()
second_class_fare = df[df['Pclass'] == 2]['Fare'].sum()
third_class_fare = df[df['Pclass'] == 3]['Fare'].sum()
total_fares = [first_class_fare, second_class_fare, third_class_fare]
print("The total fare from first class, second class and third class: ",total_fares)
```

Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.

```python
total_fare = df['Fare'].sum()
first_class_proportion = first_class_fare / total_fare
second_class_proportion = second_class_fare / total_fare
third_class_proportion = third_class_fare / total_fare

print("The proportion of fare revenue by class: ",first_class_proportion, second_class_proportion, third_class_proportion)
```

What percent of the total fare revenue came from each passenger class?

```python
first_class_revenue = first_class_proportion * 100
second_class_revenue = second_class_proportion * 100
third_class_revenue = third_class_proportion * 100
print("The total revenue from first class is: ", first_class_revenue)
print("The total revenue from second class is: ", second_class_revenue)
print("The total revenue from third class is: ", third_class_revenue)
```

# **What percent of the total number of passengers on the Titanic belonged to each age group(e.g., child, adult, senior)?**

Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age
18{64), and "senior" (age 65 and above).

```python
def age_group(age):
    if age < 18:
        return 'child'
    elif age >= 18 and age < 65:
        return 'adult'
    else:
        return 'senior'

df['Age_group'] = df['Age'].apply(age_group)
df[['Name', 'Age', 'Age_group']]
```

Calculate the total number of passengers on the Titanic.

```python
total_passengers = len(df)
print("The total number of passenger on the titanic:", total_passengers)
```

Count the number of passengers in each age group.

```python
age_group_counts = df['Age_group'].value_counts()
print(age_group_counts)
```

Divide the count of each age group by the total number of passengers to get the proportion of passengers
in each age group.

```python
age_group_proportion = age_group_counts / total_passengers
print(age_group_proportion)
```

Display the proportion as a percentage.

```python
age_group_percentage = age_group_proportion * 100
print(age_group_percentage)
```
